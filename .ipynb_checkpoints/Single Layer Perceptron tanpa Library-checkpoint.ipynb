```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
# Load dataset
data_x     = np.array([[0,0],
                       [0,1],
                       [1,0],
                       [1,1]])
data_y     = np.array([0,1,1,1])
```

```python
# Function
def sigmoid(x):
  return 1/(1+np.exp(-x))
sig = np.vectorize(sigmoid)
```

```python
#-------------------------------------
#- inisiasi

input       = data_x
target      = data_y
num_input   = 2
num_output  = 1

#Layer Input-->Layer Output
w_output = np.random.uniform(low=-3, high=1, size=(num_input,num_output))
b_output = np.random.uniform(low=-3, high=1, size=(num_output))

# Loss dan Akurasi
loss_values = []
acc_values  = []

lr     = 0.1
epochs = 500
```

```python
print(w_output.shape)
```

```python
for epoch in range(epochs):
  MSE = 0
  new_target = np.zeros(len(target))
  
  for idx, inp in enumerate(input):
    # A. Feedforward
    # Outpur Layer
    o_output  = np.matmul(input[idx], w_output) + b_output
    o_output  = sig(o_output)
    # Menghitung Error
    error           = target[idx] - o_output
    MSE             = MSE + (error*error)
    new_target[idx] = o_output.round()
    
    # B. Backpropagation
    # Outpur Layer
    w_output = w_output + (lr * ((error * o_output * (1 - o_output))* input[idx][np.newaxis].T))
    b_output = b_output + (lr * ((error * o_output * (1 - o_output))))
  
  MSE = MSE/len(target)
  acc = 1 - (np.sum(np.absolute(target-new_target))/len(target))
  loss_values.append(MSE)
  acc_values.append(acc)
  print("epoch : ",epoch," | error : ",MSE," | acc : ",acc)
```

```python
plot1 = plt.figure(1)
plt.plot(loss_values)
plt.title("Loss Values")
plt.show()

plot2 = plt.figure(2)
plt.plot(acc_values)
plt.title("Accuracy Values")
plt.show()
```

```python
def predict(data):
    HasilPredict = []
    for i in data:
      # Outpur Layer
      o_output  = np.matmul(i, w_output) + b_output
      o_output  = sig(o_output)
      HasilPredict.append(round(float(o_output)))
    return HasilPredict
```

```python
prediksi = predict([[0,1]])
print(prediksi)
#e is 0
#p is 1
for i in prediksi :
  if i == 1 :
    print('Result : 1')
  elif i == 0 :
    print('Result : 0')
```
